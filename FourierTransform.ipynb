Let consider an integrable function $u(x)=e^{-x^2}$. 
And let be $F$ the Fourier transformation of $u$

## F expression

$F(f)= \int_{-\infty}^{+ \infty} u(x) e^{-2\pi ift}$

## Discretizing

$t_m=m*\Delta_t, m=0,1,...,n-1$

$F(f)\approx\Delta_t\sum_{m=0}^{n-1}u(t_m)e^{-2\pi ift_m}$

### Frequence $f_k$
$f_k=k*\frac{1}{n\Delta_t}$

$f_kt_m=\frac{mk}{n}$ 

$F(f)\approx\Delta_t\sum_{m=0}^{n-1}u(t_m)e^{-2\pi i\frac{mk}{n}}$

$F(f)\approx\Delta_t\sum_{m=0}^{n-1}u(t_m)\cos (-2\pi i\frac{mk}{n})+i\Delta_t\sum_{m=0}^{n-1}u(t_m)\sin (-2\pi i\frac{mk}{n})$


## Implementing

### Modules

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

### Initial Values

```python
n=10
m=np.linspace(0,n,n, dtype=int)
delta_t=0.1
t_m=np.linspace(0,m*delta_t,n)
```

### Code

```python
u_tm=np.exp(-t_m**2)
plt.subplot()
plt.plot(m,u_tm, label="Funct u")
plt.legend()
```

```python
k=1
## fourier sum
re = delta_t*sum(np.exp(t_m**2)*np.cos(-2*math.pi*m*k/n))
im = delta_t*sum(np.exp(t_m**2)*np.sin(-2*math.pi*m*k/n))

z=re+1j*im
plt.plot(m, np.real(z), label="real")
plt.plot(m, np.imag(z), label="imaginary")
#plt.plot(m, np.abs(z), label="mod")
plt.legend()
```

```python
k=1
## fourier sum
re = delta_t*sum(np.exp(t_m**2)*np.cos(-2*math.pi*m*k/n))
im = delta_t*sum(np.exp(t_m**2)*np.sin(-2*math.pi*m*k/n))

z=re+1j*im
#plt.plot(t_m, np.real(z), label="real")
#plt.plot(t_m, np.imag(z), label="imaginary")
plt.plot(t_m, np.abs(z), label="mod")
plt.legend()
```

### Representation by changing $k$

$k=0,1,2,...,K$


```python
x = []
y = []
K=10
for k in range(0, K):
    re = delta_t*sum(np.exp(t_m**2)*np.cos(-2*math.pi*m*k/n))
    im = delta_t*sum(np.exp(t_m**2)*np.sin(-2*math.pi*m*k/n))
    x.append(re)
    y.append(im)
plt.scatter(x, y, label="star", marker="*", color="green")
plt.xlabel('real axis')
plt.ylabel('imaginary axis')
plt.title('complex numbers')
plt.legend()
plt.show()
```
