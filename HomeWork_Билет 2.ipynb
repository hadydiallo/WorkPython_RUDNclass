## 2. –í—ã–≤–µ—Å—Ç–∏ –≥—Ä–∞—Ñ–∏–∫–∏ —Ñ—É–Ω–∫—Ü–∏–∏ –∏ –µ—ë –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π $e^x \cos{x}$. –ù–∞–π—Ç–∏ min, max,  —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –Ω–∞ –æ—Ç—Ä–µ–∑–∫–µ $[0;\pi/2]$.

$f(x)=e^x \cos{x}$ , $x\in [0;\pi/2] $

$f'(x)=e^x (\cos{x}-\sin{x})$


#### –≥—Ä–∞—Ñ–∏–∫–∏ —Ñ—É–Ω–∫—Ü–∏–∏ –∏ –µ—ë –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π

```python
## pip install ggplot
```

```python
from math import pi, cos, sin, exp
import matplotlib.pyplot as plt
#import ggplot ## make the graphic more beautiful

x=np.linspace(0,pi/2,10)
def f(x):
    return exp(x)*cos(x)
def ff(x) :
    return exp(x)*(cos(x)-sin(x))
f_x=np.zeros(len(x))
ff_x=np.zeros(len(x))
for i in range(len(x)):
    f_x[i]=f(x[i])
    ff_x[i]=ff(x[i])

fig, ax = plt.subplots()
ax.plot(x,f_x, label='f(x)')
ax.plot(x,ff_x, label="f'(x)")
ax.set_xlabel("x")
ax.set_ylabel("functions")
ax.set_title("—Ñ—É–Ω–∫—Ü–∏–∏ –∏ –µ—ë –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π f(x)")
ax.legend()
#fig.show()
```

#### –ù–∞–π—Ç–∏ min, max, —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –Ω–∞ –æ—Ç—Ä–µ–∑–∫–µ  [0;ùúã/2] 

```python
Min=np.round(np.min(f_x),5)
Max=np.round(np.max(f_x),5)
Mean=np.round(np.mean(f_x),5)
print("min= ",Min," Max= ",Max," —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ= ",Mean)
```

```python
data_path = "comparison-ww-monthly-201902-202002.csv"
types = ['a25','f4', 'f4', 'f4']
data = np.genfromtxt(data_path, dtype=types, delimiter=',', names=True)
data
```

#### –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏.

```python
# Plot the data
fig, ax = plt.subplots()
ax.plot( data['Date'], data['Mobile'], label='Mobile')
ax.plot( data['Date'], data['Desktop'], label='Desktop')
ax.plot( data['Date'], data['Tablet'], label='Tablet')
ax.set_xlabel("Date")
ax.set_ylabel("Frequency")
ax.set_title("Comparaison of using tools from feb 2019 to feb 2020")
ax.legend()
fig.show()
def comp_graph(a) :
    fig, ax = plt.subplots()
    ax.plot(data['Date'],data[a], 'b', label=a)
    ax.set_xlabel("Date")
    ax.set_ylabel("Frequency")
    ax.set_title("Using "+a+" from feb 2019 to feb 2020")
    ax.legend()
    fig.show()
```

```python
comp_graph('Mobile')
comp_graph('Desktop')
comp_graph('Tablet')
```
